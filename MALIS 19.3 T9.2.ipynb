## Datenanalyse der offenen Daten der Seattle Library


Die URL definiert in welcher Größe die Datenmengen abgerufen werden. So kann man die URL auf Zeilen (standartmäßig 1000) oder auf Jahre begrenzen. Für die Fragen wurde ist ein Limit von 10.000 Zeilen oder ein Jahr mit 10.000 Zeilen genommen worden. Der benutzte Rechner musste bei einem Versuch die Datenmenge eines Jahres zu laden neu gestartet werden. Folgende fünf Fragen sollen mittels der Datenanalyse beantwortet werden:

# Frage "Welcher Titel wurde 2018 am häufigsten entliehen?"

Hierfür wird zunächst pandas nach Anleitung heruntergeladen. Dann werden die Daten importiert. Um den häufigst entliehenen Titel zu finden, wird nach der maximalen Anzahl an Checkouts gesucht und diese Zeile wird angezeigt.

Zu beachten: die Datenmenge besteht, durch die Begrenzung der URL, nur aus Ausleihen im Dezember 2018.

```python
import urllib.request
import pandas as pd
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=10000"
ceins = "Checkouts.csv"
```

```python
# Download der Daten
urllib.request.urlretrieve(url, ceins)
```

```python
checkout = pd.read_csv(ceins)
```

```python
# Anzeigen der Tabelle zur Kontrolle 
checkout
```

```python
# Anzeige aller Checkouts
checkout.checkouts
```

```python

checkout["checkouts"] >= 5
```

```python
# Annährung an die häufigste Ausleihe und Ermittlung von Ausleihzahlen
checkout[checkout["checkouts"] >= 103]
```

```python
# checkmax wird definiert, um im Folgenden die maximale Checkoutanzahl mit ganzer Zeile anzuzeigen
checkmax = checkout.checkouts.max()
```

```python
checkmax
```

```python
# Anzeigen der Zeile mit höchster Checkout-Anzahl
checkout[checkout.checkouts >= checkmax]
```
