## Datenanalyse der offenen Daten der Seattle Library


Die URL definiert in welcher Größe die Datenmengen abgerufen werden. So kann man die URL auf Zeilen (standartmäßig 1000) oder auf Jahre begrenzen. Für die Fragen wurde ist ein Limit von 3.000.000 Zeilen genommen, damit sind alle Checkouts des Jahres 2018 enthalten. Bei dieser Frage (Frage 5) wurde zusätzlich noch alle Checkouts des Jahres 2005 geladen.

# Frage 5 "Sieht man von 2005 bis 2018 eine Veränderung der UsageClass? Wie ist das Verhältnis der Ausleihen zwischen digitalen und physischen Medien?"

Hierfür wird zunächst pandas nach Anleitung heruntergeladen. Dann werden die Daten importiert, bei dieser Frage wurden die Checkots von 2005 und 2018 zum Vergleich geladen.Es wurde nach digitalen und physischen Medien gefiltert. Zuletzt werden die Medien nach digitalen und physischen Medien differenziert und die Jahre 2005 und 2018 angezeigt

```python
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
```

```python
urlfuenf = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2005&$limit=3000000"
urlachtzehn = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=3000000"
fuenf = "Checkoutsfuenf.csv"
achtzehn = "Checkoutsachtzehn.csv"
```

```python
# Download der Daten
 urllib.request.urlretrieve(urlfuenf, fuenf)
 urllib.request.urlretrieve(urlachtzehn, achtzehn)
```

```python
checkoutfuenf = pd.read_csv(fuenf)
checkoutachtzehn = pd.read_csv(achtzehn)
```

```python
# Anzeige zur Kontrolle. So wird deutlich, dass das Maximum von 3.000.000 Zeilen nicht erreicht wurde.
checkoutfuenf
```

```python
checkoutachtzehn
```

```python
# Anzeige der Spaltentitel
checkoutfuenf.columns
```

```python
# Anzeige der Spaltentitel. Kontrolle der identischen Spaltentitel.
checkoutachtzehn.columns
```

```python
# Definition der Variabel und Zusammenfassung der Daten mit Hilfe der Pivot-Tabelle.
fuenfp = checkoutfuenf.pivot_table(index = "usageclass", values = "title", aggfunc="count")
```

```python
# Anzeige der Verteilung
fuenfp
```

```python
# Definition der Variabel und Zusammenfassung der Daten mit Hilfe der Pivot-Tabelle (s.o.).
achtp = checkoutachtzehn.pivot_table(index = "usageclass", values = "title", aggfunc="count")
```

```python
# Anzeige der Verteilung
achtp
```

```python
# graphische Darstellung des Jahres 2005
fuenfp.plot(kind = "bar")
```

```python
# graphische Darstellung des Jahres 2018
achtp.plot(kind = "bar")
```
