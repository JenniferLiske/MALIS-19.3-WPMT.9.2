## Datenanalyse der offenen Daten der Seattle Library


Die URL definiert in welcher Größe die Datenmengen abgerufen werden. So kann man die URL auf Zeilen (standartmäßig 1000) oder auf Jahre begrenzen. Für die Fragen wurde ist ein Limit von 3.000.000 Zeilen genommen, damit sind alle Checkouts des Jahres 2018 enthalten. 
Folgende fünf Fragen sollen mittels der Datenanalyse beantwortet werden:

# Frage 2 "Wie oft wurde Harry Potter ausgeliehen?"

Hierfür wird zunächst pandas nach Anleitung heruntergeladen. Dann werden die Daten importiert. 

```python
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=3000000"
ceins = "Checkoutsall.csv"
```

```python
# Download der Daten
# urllib.request.urlretrieve(url, ceins)
```

```python
checkout = pd.read_csv(ceins)
```

```python
# Anzeigen der Tabelle zur Kontrolle 
checkout
```

```python
print(checkout.title)
```

```python
harry = checkout.title.str.contains(pat="Harry Potter")
```

```python
harry
```

```python
harryall = checkout[harry] 
```

```python
harryall
```

```python
# NaN muss als falsch definiert werden, da sonst ein Fehler ausgegeben wird.
rowling = harryall.creator.str.contains(pat="Rowling", na = False)
```

```python
rowling
```

```python
harryrowling = harryall[rowling]
```

```python
harryrowling
```

```python
# Harry Potter wurde 14.438 mal ausgeliehen. Hiermit sind alle Titel gemeint, die Harry Potter im Namen haben und von J.K. Rowling geschrieben wurden. 
# Wie durch die Tabelle oben erkenntlich ist, fallen darunter auch Harry Potter-Bücher in nicht-englisch und mehrfach vorhandene Titel
harryrowling.sum() 
```

```python
# Die Lösung wird genuaer betrachtet. Um eine Verteilung auf die Monate zu sehen, wird aus einer Pivot-Tabelle ein Graph erzeugt
harryrowling[["title", "checkouts"]]
```

```python
harryr = harryrowling.pivot_table(index = "checkoutmonth", values = "checkouts", aggfunc="sum")
```

```python
harryr
```

```python
harryr.plot(kind = "bar")
```
